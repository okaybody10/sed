```python
import csv
def make_index_dict(label_csv):
    index_lookup = {}
    with open(label_csv, 'r') as f:
        csv_reader = csv.DictReader(f)
        line_count = 0
        for row in csv_reader:
            index_lookup[row['mid']] = row['display_name']
            line_count += 1
    return index_lookup

label_csv = '/gallery_tate/jaehyuk.sung/sed/datasets/audioset201906/Fast-Audioset-Download/csvs/class_labels_indices.csv'
index_dict = make_index_dict(label_csv)
print(index_dict)
```

```python
import json
audioset_description = '/gallery_tate/jaehyuk.sung/sed/audioset_descript.json'
with open(audioset_description, "r") as fp :
    audioset_json = json.load(fp)
print(audioset_json)
```

```python
audioset_label = list(audioset_json.keys())
label_dscription = list(audioset_json.values())
```

```python
from sentence_transformers import SentenceTransformer, util
import torch

model = SentenceTransformer("sentence-transformers/all-mpnet-base-v2")

# Two lists of sentences
sentences1 = [
    *label_dscription
]

prompt = ['exclamation', 'sound effect', 'music', 'speech', 'background noise']

# Compute embedding for both lists
embeddings1 = model.encode(sentences1, convert_to_tensor=True)
embeddings2 = model.encode(prompt, convert_to_tensor=True)

# Compute cosine-similarities
cosine_scores = util.cos_sim(embeddings1, embeddings2)
# print(cosine_scores)
convert_label =  [prompt[i] for i in torch.argmax(cosine_scores, axis=1)]
# print(convert_label, len(convert_label))
```

```python
convert_label_dict = dict(zip(audioset_label, convert_label))
print(prompt)
print(convert_label_dict)
```

```python
def convert_labels(label_collect) :
    return list(set([convert_label_dict[index_dict[i]] for i in label_collect]))
# print(data['id_MdYXznF3Eac']['labels'], convert_labels(data['id_MdYXznF3Eac']['labels']), sep='\n')
for key, value in data.items() :
    data[key]['labels'] = convert_labels(data[key]['tags'])
```
